```python
#load data set
from datasets import load_dataset
import pandas as pd
#https://huggingface.co/datasets/MongoDB/embedded_movies

dataset = load_dataset("AIatMongoDB/embedded_movies")

dataset_data_frame = pd.DataFrame(dataset['train'])
dataset_data_frame.head(5)
```

```python
print("\nNumber of missing values in each column before cleaning")
print(dataset_data_frame.isnull().sum())
```

```python
## clearly, there are 48 entries where the plot is missing of the movie so we want to clean up this data..
## remove data where plot column is missing.
dataset_data_frame = dataset_data_frame.dropna(subset=['plot'])
print("\nNumber of missing values in each column after cleaning ")
print(dataset_data_frame.isnull().sum())
```

```python
import openai
from google.colab import userdata
openai.api_key = "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"
EMBEDDING_MODEL = "text-embedding-3-small"

def get_embedding(text):
  """Generate an embedding for the given text OPEN AI's API"""
  #check for valid input

  if not text or not isinstance(text,str):
    return None

  try:
    #Call OPen AI's API to get the embedding
    embedding = openai.embeddings.create(input=text, model= EMBEDDING_MODEL).data[0].embedding
    return embedding

  except Exception as e:
    print(f"Error in get_embedding: {e}")
    return None

    # we are running embed on just the plot. We are setting a new column as plot embedding optomised

dataset_data_frame["plot_embedding_optomised"] = dataset_data_frame['plot'].apply(get_embedding)
dataset_data_frame.head()
```

```python
# Now lets write the function that will perform vector search on user queries

def vector_search(user_query, collection):

  # Generate embedding for the user query! So we can compare the query to

  query_embedding = get_embedding(user_query)

  if query_embedding is None:
    return "Invalid query or embedding generation failed :/"

  # defining vector search pipeline

  pipeline = [
      {
          "$vectorSearch": {
              "index" : "vector_index",
              "queryVector" : query_embedding,
              "path": "plot_embedding_optimised",
              "numCandidates": 150,
              "limit": 5 # return top 5 matches
          }
      },
      {
          "$project": {
              "_id": 0, # Exlude the id field
              "plot": 1, # Include the plot field
              "title": 1 ,
              "genres": 1 ,
              "score": {
                  "$meta": "vectorSearchScore" # Include the search score

          }

      }
      }
  ]

  #Execute the seach
  results = collection.aggregate(pipeline)
  return list(results)





```

```python
# now lets handle the user query

def handle_user_query(query, collection):
  get_knowledge = vector_search(query, collection)
  search_result = ''
  for result in get_knowledge:
    search_result += f"Title {result.get('title', 'N/A')}, Plot {result.get('plot', 'N/A')}\n"

  completion = openai.chat.completions.create(
      model = "gpt-3.5-turbo",
      messages = [
          {"role": "system", "content": "You are a movie recommendation system."},
          {"role": "user", "content": "Answer this user query: " + query + " with the following context: " + search_result}


      ]

  )

  return (completion.choices[0].message.content), search_result

```

```python
# FINally, lets ask a querry:

query = "What are the best action movies?"
response, source_information = handle_user_query(query, collection)
print(f"Response : {response}")


```
